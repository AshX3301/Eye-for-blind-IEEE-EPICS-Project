```python
import os
image_path= "data/super_data"
path_list = []
for cur_dir,dirs,files in os.walk("."):
  for f in files:
    if f.endswith('.jpg'):
      file_loc =  image_path + '/' + f 
      path_list.append(file_loc + '\n')
```

```python
path_list
```

```python
test = path_list[:int(len(path_list)* 0.20)]
```

```python
train = path_list[int(len(path_list)* 0.20):]
```

```python
len(test)
```

```python
with open("train.txt","w") as t:
  for i in train:
    t.write(i)
```

```python
with open("test.txt","w") as t:
  for i in test:
    t.write(i)
```

```python
with open("classes.txt", "w") as file:
    file.write("colgate\n")
    file.write("avt tea powder\n")
    file.write("milkybar\n")
    file.write("oreo\n")
    file.write("vim\n")
```

```python
%cd 'data/super_data'
```

```python
i = 0

with open('classes.names','w') as cls,\
   open('classes.txt','r') as text:


   for l in text:
     cls.write(l)

     i+=1
```

```python

with open('.' + '/' + 'image_data.data','w') as data:
  data.write('classes = ' + str(i) + '\n') 

  data.write('train = ' + image_path + '/' + 'train.txt' + '\n')

  data.write('valid = ' + image_path + '/' + 'test.txt' + '\n')

  data.write('names = ' + image_path + '/' + 'classes.names' + '\n')

  data.write('backup = backup')
```
